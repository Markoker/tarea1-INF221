```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
csvs = ["selection_sort.csv", "merge_sort.csv", "quick_sort.csv", "funcion_implementada.csv"]
folders = ["sorted", "rev_sorted", "unsorted", "nearly_sorted"]
folder_name = ["ordenados", "ordenados inversamente", "desordenados", "casi ordenados"]
nombres = ["Selection sort", "Merge sort", "Quick sort", "C++ sort"]
colores = ["#fd7f6f", "#ffb55a", "#7eb0d5", "#bd7ebe"]
```

```python
GREY10 = "#1a1a1a"
GREY30 = "#4d4d4d"
GREY40 = "#666666"
GREY50 = "#7f7f7f"
GREY60 = "#999999"
GREY75 = "#bfbfbf"
GREY91 = "#e8e8e8"
GREY98 = "#fafafa"
```

# Comparación de algoritmos de ordenamiento en arreglos casi ordenados

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Background color
fig.patch.set_facecolor(GREY98)
ax.set_facecolor(GREY98)

plt.grid(False)

VLINES = np.arange(1,6,1)

# Vertical lines used as scale reference
for h in VLINES:
    ax.axvline(h, color=GREY91, lw=0.6, zorder=0)

# Horizontal lines
ax.hlines(y=np.arange(2, 11), xmin=1, xmax=5, color=GREY91, lw=0.6)

offsets = [0,0.15,0,-0.12]

PAD = 0.1
for file in range(len(csvs)):
    data = pd.read_csv("/home/marcor/CLionProjects/tarea1-INF221/ordenamiento/data/nearly_sorted/" + csvs[file])
    
    data["Tiempo"] = np.log10(data["Tiempo"])
    data["Tamaño"] = np.log10(data["Tamaño"])
    
    x_start = data["Tamaño"].iloc[-1]
    x_end = x_start + 3*PAD
    y_start = data["Tiempo"].iloc[-1]
    y_end = data["Tiempo"].iloc[-1] + offsets[file]
    
    ax.plot(
        [x_start, (x_start + x_end - PAD) / 2 , x_end - 0.3*PAD], 
        [y_start, y_end, y_end], 
        color=colores[file], 
        alpha=0.2, 
        ls="dashed"
    )
    
    ax.text(
        x_end, 
        y_end, 
        nombres[file], 
        color=colores[file], 
        fontsize=10, 
        va="center"
    )
    
    ax.plot(data["Tamaño"], data["Tiempo"], color=colores[file])
        
ax.set_xticks([x for x in np.arange(1, 6, 1)])
ax.set_xticklabels(
    [x for x in np.arange(1, 6, 1)], 
    fontsize=10,
    weight=500,
    color=GREY40
)

ax.set_xlim(1, 6.1)
plt.xlabel("Tamaño del arreglo (n) [log10]")
plt.ylabel("Tiempo (ns) [log10]")
plt.title("Comparación de algoritmos de ordenamiento en arreglos casi ordenados")
plt.savefig("/home/marcor/CLionProjects/tarea1-INF221/ordenamiento/graficos/comparacion_nearly_sorted.png",
            dpi=300, bbox_inches="tight")
```

# Comparacion de la suma total de tiempo

```python
data_b = []

fig, ax = plt.subplots(figsize=(10, 6))

# Background color
fig.patch.set_facecolor(GREY98)
ax.set_facecolor(GREY98)

VLINES = np.arange(0,13,1)

# Vertical lines used as scale reference
for h in VLINES:
    ax.axvline(h, color=GREY91, lw=0.6, zorder=0)

plt.grid(False)

for file in range(len(csvs)):
    data = pd.read_csv("/home/marcor/CLionProjects/tarea1-INF221/ordenamiento/data/nearly_sorted/" + csvs[file])
    
    data_b.append(data["Tiempo"].sum())
    
data_b = np.log10(data_b)

# Sort data_b and change the order of the names, colors and bars
data_b, new_nombres, new_colores = zip(*sorted(zip(data_b, nombres, colores)))
    
bars = plt.barh(new_nombres, data_b, color=new_colores)
    
for bar in bars:
    ax.text(0 + 0.1,  # Posición en el eje x (justo después de la barra)
            bar.get_y() + bar.get_height()/2,  # Posición en el eje y (centrado en la barra)
            # Agrega comas para separar miles y convierte a string
            f'{int(10**(bar.get_width())):,}',
            color=GREY98,
            va='center')  # Alinear verticalmente al centro
    
#ax.set_xlim(1, 6.1)
plt.xlabel("Tamaño del arreglo (n) [log10]")
plt.ylabel("Tiempo (ns) [log10]")
plt.title("Comparación suma total de tiempo para algoritmos de ordenamiento en arreglos casi ordenados")
plt.savefig("/home/marcor/CLionProjects/tarea1-INF221/ordenamiento/graficos/comparacion_total_nearly_sorted.png",
            dpi=300, bbox_inches="tight")
```
